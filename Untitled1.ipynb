**ЗАДАЧА.** Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
penguins.loc[penguins['bill_length_mm'] >= 42, 'hight_group'] = 'high'
penguins.loc[(penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] < 42), 'hight_group'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'hight_group'] = 'low'
```

```python
penguins
```

```python
sns.histplot(data = penguins,x ='flipper_length_mm', hue="hight_group",multiple='stack')
```
